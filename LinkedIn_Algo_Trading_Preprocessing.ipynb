```python
import pandas as pd
```

```python
df_microsoft = pd.read_csv('MSFT.csv', parse_dates=['Date'], index_col=0)
df_microsoft.head()
```

```python
df_microsoft_linkedin = df_microsoft.loc['2016-12-08':,:].copy()
```

```python
df_microsoft_linkedin['change_tomorrow'] = df_microsoft_linkedin.Close.pct_change(-1) * 100 * -1
```

```python
df_microsoft_linkedin
```

```python
df_microsoft_linkedin = df_microsoft_linkedin.dropna().copy()
```

```python
import numpy as np
```

```python
df_microsoft_linkedin['change_direction_tomorrow'] = np.where(df_microsoft_linkedin.change_tomorrow > 0, 'UP', 'DOWN')
```

```python
df_microsoft_linkedin.Close.plot()
```

```python
df_microsoft_linkedin.to_excel('microsoft_linkedin_processed.xlsx')
```

```python
df_microsoft_linkedin.to_csv('microsoft_linkedin_processed.csv')
```
