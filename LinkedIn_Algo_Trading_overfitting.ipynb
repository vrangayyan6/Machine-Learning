```python
import yfinance as yf
```

```python
ticker = "AAPL"
df = yf.download(ticker)
df
```

```python
df = df.loc['2023-01-01':].copy()
df
```

Add % change

```python
df['change tomorrow'] = df['Adj Close'].pct_change(-1) * -1 * 100
df
```

remove rows with missing data

```python
df = df.dropna().copy()
df
```

```python
X = df.drop(columns=['change tomorrow'])
y = df['change tomorrow']
```

Train test split

```python
n_days = len(df.index)
n_days
```

```python
n_days_train = round(n_days * 0.7)
n_days_train
```

```python
X_train = X.iloc[:n_days_train]
y_train = y.iloc[:n_days_train]
X_test = X.iloc[n_days_train:]
y_test = y.iloc[n_days_train:]
```

Fit model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_dt = DecisionTreeRegressor(max_depth=20, random_state=40)
model_dt.fit(X_train, y_train)
```

Evaluate model

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred_train = model_dt.predict(X_train)
mean_squared_error(y_train, y_pred_train)
```

```python
y_pred_test = model_dt.predict(X_test)
mean_squared_error(y_test, y_pred_test)
```

Backtesting

```python
from backtesting import Backtest, Strategy
```

Create Strategy

```python
class RegressionStrategy(Strategy):
  limit_buy = 1
  limit_sell = -1

  def init(self):
      self.model = DecisionTreeRegressor(max_depth=20, random_state=40)
      self.model.fit(X_train, y_train)
      self.already_bought = False

  def next(self):
      inputs_today = self.data.df.iloc[[-1],:]
      forecast_tomorrow = self.model.predict(inputs_today)[0]

      if forecast_tomorrow > self.limit_buy and not self.already_bought:
          self.buy()
          self.already_bought = True
      elif forecast_tomorrow < self.limit_sell and self.already_bought:
          self.sell()
          self.already_bought = False
      else:
          pass
```

Run backtest

```python
bt_test = Backtest(X_test, RegressionStrategy, cash=10000, commission=0.002, exclusive_orders=True)
```

```python
result_test = bt_test.run()

df_result_test = result_test.to_frame(name='Values').loc[:'Return [%]']
df_result_test
```

Run backtest on training data

```python
bt_train = Backtest(X_train, RegressionStrategy, cash=10000, commission=0.002, exclusive_orders=True)
```

```python
result_train = bt_train.run()

df_result_train = result_train.to_frame(name='Values').loc[:'Return [%]']
df_result_train
```

```python
import pandas as pd
```

```python
df_result = pd.concat([df_result_train, df_result_test], axis=1)
df_result
```

Plot backtest reports

```python
bt_test.plot(filename='regression_testing.html')
bt_train.plot(filename='regression_training.html')
```
